```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_pos = [1, 2, 3, 4, 5, 6, 7]
y_pos = [2, 3, 4, 5, 6, 7, 8]

x_neg = [1, 2, 3, 4, 5, 6, 7]
y_neg = [10, 9, 8, 7, 6, 5 , 4]

x_zero = [2,5,1,3,7,6,9,3,6]
y_zero = [1,3,7,5,9,2,3,7,6]
```

```python
# Correlations
corr_pos= np.corrcoef(x_pos, y_pos)[0, 1]
corr_neg= np.corrcoef(x_neg, y_neg)[0, 1]
corr_zero = np.corrcoef(x_zero, y_zero)[0, 1]
```

```python
# Plotting

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(x_pos, y_pos, color='blue')
plt.title("Positive Correlation")

plt.subplot(1, 3, 2)
plt.scatter(x_neg, y_neg, color='red')
plt.title("Negative Correlation")

plt.subplot(1, 3, 3)
plt.scatter(x_zero, y_zero, color='green')
plt.title("Zero Correlation")

plt.tight_layout()
plt.show()
```
